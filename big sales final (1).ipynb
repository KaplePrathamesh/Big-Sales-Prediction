```python
y =sale['Item_Outlet_Sales']
y
```

```python
X =sale[[ 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility',
       'Item_Type', 'Item_MRP', 'Outlet_Identifier',
       'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type',
       'Outlet_Type']]
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_std= sale[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
```

```python
X_std= sc.fit_transform(X_std)
X_std
```

```python
X[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']] = pd.DataFrame(X_std, columns= [['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1 , random_state=2529)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=2529)
```

```python
y_train=y_train.astype(str)
```

```python
X_train= pd.get_dummies(X_train)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred= rfr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
